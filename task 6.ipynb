# Problem Statement 1:Smart Parking Lot Management System


```python
def smart_parking_lot_management_system():
    i=1
    max_parking_limit = 50
    parking= []
    while(i!=0):
        parking_status = input("ENTER YOUR PARKING STATUS (in/out)").lower()
        
        if(parking_status == "in"):
            if(parking.count('in')>=max_parking_limit):
                print("PARKING LIMIT EXCEEDS: ",len(parking))
                status = "UNAVAILABLE"
                print(status)
            else:
                parking.append(parking_status)
                print("CURRENT COUNT OF PARKING : ",parking.count('in'))
                status = "AVAILABLE"
                print(status)
        else:
            for i in range (len(parking)) :
                if parking[i]== "in":
                    parking[i]='out'
                    break
            
                
        parking_vacancy = parking.count('in')   
        print("PARKING CONDITION: ",parking)
        print("PARKING VACANCY: ",max_parking_limit-parking_vacancy, "\n")
        i=int(input("DO YOU WANTS TO PARK ANOTHER VEICHEL (1/0)"))
    print("\n DAILY STATS : \n" )
    print("NUMBER OF VEICHEL in : ",parking.count('in'))
    print("NUMBER OF VEICHEL out: ",parking.count('out'))
    print("TOTAL: ", parking.count('in')+ parking.count('out'))

smart_parking_lot_management_system()
        
```

#  Problem Statement 2:Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic_level, weather):
    base_speed = 30
    
    base_time = (distance / base_speed) * 60
    
    traffic_multipliers = {
        "low": 1.0,
        "medium": 1.2,
        "high": 1.4
    }
    
    weather_multipliers = {
        "clear": 1.0,
        "rainy": 1.25,
        "stormy": 1.5
    }
    
    traffic_level = traffic_level.lower()
    weather = weather.lower()
    
    if traffic_level not in traffic_multipliers:
        return "Invalid traffic level entered."
    if weather not in weather_multipliers:
        return "Invalid weather condition entered."
    
    final_time = base_time * traffic_multipliers[traffic_level] * weather_multipliers[weather]
    
    return round(final_time)
    
distance = float(input("Enter Distance (km): "))
traffic = input("Enter Traffic Level (Low/Medium/High): ")
weather = input("Enter Weather (Clear/Rainy/Stormy): ")

eta = estimate_delivery_time(distance, traffic, weather)

print(f"Estimated Delivery Time: {eta} minutes")
```

#  Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def movie_theater_seat_occupancy_analyzer(theater_capacity,seat_list):
    occupied_seat_count = seat_list.count(1)
    percentage_of_occupancy = (occupied_seat_count/theater_capacity)*100
    print("OCCUPANCY: ",percentage_of_occupancy)
    
    if percentage_of_occupancy >= 0 and percentage_of_occupancy<=50:
        print("theater occupancy is below 50 %")
    elif percentage_of_occupancy > 50 and percentage_of_occupancy<75:
        print("about to full")
    else:
        print("alomst full")

theater_capacity = 10
seat_list = []
for i in range (1,theater_capacity+1):
    print("seat no: ",i)
    seat_status=int(input("status of seat (1/0)"))
    seat_list.append(seat_status)

print(seat_list)
movie_theater_seat_occupancy_analyzer(theater_capacity,seat_list)    
    
```

#  Problem Statement 4:Cloud Server Load Classification System


```python
def cloud_server_load_classification_system(cpu_usage):
    summ = 0
    for usage in cpu_usage:
        summ = summ + int(usage)
        
    average = (summ /len(cpu_usage))
    print("Average Usage= ",average)
    if average <=50:
        print("good going")
    elif(average >50 and average <=70):
        print("normal going")
    else:
        print("Warning") 

usage_list=[]
while True:
    cpu_usage = input("enter the cpu usage")
    if cpu_usage.strip()=="":
        break
    try:
        num = int(cpu_usage)
        usage_list.append(num)
    except ValueError:
        print("plz enter valid digit")

print("Usage List in (%)",usage_list)
cloud_server_load_classification_system(usage_list)
```

#  Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def monitor_resource_usage(resource_usage):
 
    usage_limits = {
        "Projector": 6,
        "AC": 8,
        "Lights": 6
    }
    
    overused = []
    
    # Check usage
    for resource, hours in resource_usage.items():
        if hours > usage_limits.get(resource, 0):
            overused.append(resource)
    
    energy_alert = "Yes" if overused else "No"
    
    return overused, energy_alert

resource_usage = {}

projector_hours = float(input("Enter Projector usage hours: "))
ac_hours = float(input("Enter AC usage hours: "))
lights_hours = float(input("Enter Lights usage hours: "))

resource_usage["Projector"] = projector_hours
resource_usage["AC"] = ac_hours
resource_usage["Lights"] = lights_hours

overused_resources, alert = monitor_resource_usage(resource_usage)

if overused_resources:
    print("Overused Resources:", ", ".join(overused_resources))
else:
    print("Overused Resources: None")

print("Energy Alert:", alert)
```

#  Problem Statement 6: Online Event Registration Capacity Controller


```python
def online_event_regesrtation_capacity_controller(event_capacity,regestrations):
    if(regestrations == event_capacity):
        confirmed = regestrations
        waiting  = 0
        regestration_status = "Closed"
    elif (regestrations > event_capacity):
        confirmed = event_capacity
        waiting = regestrations-event_capacity 
        regestration_status = "Closed"
    else:
        confirmed = regestrations
        waiting =0
        regestration_status = "Opened"

    return f"\n Confirmed Regestrations = {confirmed} \n Waiting List {waiting} \n Regestration Status {regestration_status}"
    

event_capacity = int(input("enter event capacity"))
regestrations = int (input("enter event regestration"))
result = online_event_regesrtation_capacity_controller(event_capacity,regestrations)
print(result)
```
