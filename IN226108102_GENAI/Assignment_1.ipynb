### Q4 Stemming vs Lemmatization
**ANS:-** 
- Stemming: Cuts words to root form (may not be meaningful)
-     Example: "running" → "run", "better" → "bet"
- Lemmatization: Converts words to meaningful base form using vocabulary
-     Example: "better" → "good"



# ✅ Task 2: Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    try:
        if not text or not isinstance(text, str):
            return [], ""
        
        # Remove URLs & emails
        text = re.sub(r"http\S+|www\S+|https\S+", "", text)
        text = re.sub(r"\S+@\S+", "", text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Convert to lowercase
        text = text.lower()

        # Handle repeated characters (e.g., sooo → so)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove special characters & emojis
        text = re.sub(r'[^\w\s]', '', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        tokens = text.split()

        # Remove short tokens except 'no', 'not'
        tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

        cleaned_sentence = " ".join(tokens)

        return tokens, cleaned_sentence

    except Exception as e:
        print("Error:", e)
        return [], ""
```

# ✅ Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in test_sentences:
    tokens, cleaned = preprocess_text(text)
    print("\nOriginal:", text)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
```

# ✅ Task 4: Token Analytics

```python
def token_analysis(tokens):
    if not tokens:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total
    
    return total, unique, avg_len


for text in test_sentences:
    tokens, cleaned = preprocess_text(text)
    total, unique, avg = token_analysis(tokens)

    print("\nSentence:", text)
    print(f"Total Tokens: {total}, Unique Tokens: {unique}, Avg Length: {avg:.2f}")
```

# ✅ Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

# ✅ Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
