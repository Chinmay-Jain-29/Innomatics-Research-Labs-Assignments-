# Problem 1: Mobile Recharge Validation System


```python
def mobile_recharge_validation(amount, plan):

    if amount<=50:
        print ("Invalid Recharge : Amount must be greater than 50")
        return False
    elif plan not in [ 199, 299, 399, 599]:
        print ("Invalid Recharge : plan is invalid")
        return False

    else:
        print ("RECHARGE SUCCESSFUL..")
        return True
   
while(True):
    try:
        
        amount = int(input ("enter amount "))
        plan = int (input("enter plan (199, 299, 399, 599)"))
        if mobile_recharge_validation(amount,plan):
            break;
        else:
            print ("TRY AGAIN..")
            
    except ValueError:
          print("âŒ Please enter a valid number.\n")
```

# Problem 2: Inventory Reorder Alert System


```python
records = {}

num_of_items=int (input("enter the number of items "))
for i in range(num_of_items):
    key = input("enter the inventory items ..")
    value =int (input("enter the values for that items "))
    records[key] = value

print (records)

for item in records:
    if records[item] < 15:
        print(item, "invalid stock")
    else:
        print(item, "OK")
```

#  Problem 3: Student Result Processing System


```python
def Student_result_system():
    marklist = []
    num_of_subjects = int (input("Enter the number of subjects"))

    for i in range (num_of_subjects):
        marks = int(input("Enter your marks.."))
        marklist.append(marks)
    print(marklist)

    sum = 0
    for marks in marklist:
        sum = marks+sum
    avg = sum/num_of_subjects
    print (avg)

    if avg >= 50:
        return "PASS.."
    else:
        return "FAIL"
    
result = Student_result_system()
print(result)
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
def cab_fair_estimation(km,time):
    base_price = 50
    fair = base_price+km*12
    if time in ["7am","8am","9am","10am","11am", "5pm","6pm","7pm","8pm","9pm"]:
        fair = fair * 1.25
    return fair


while True:
    
    distance = int(input("enter the distance to travel"))
    time = input("Enter your travelling time (am/pm)").lower()
    
    result = cab_fair_estimation(distance, time)
    print(result)
    repeat = input("DO YOU WANTS TO CALCULATE MORE (Y/N)").upper()
    if repeat == 'N':
        break
    
```

# Problem 5: Employee Attendance Eligibility Checker


```python
def  employee_attendance_eligibility_checker(attendence):
    present_days = 0
    total_days=0

    for i in attendence:
        total_days+=1
        if i == 'P':
            present_days +=1
        else:
            present_days +=0

    attendence_percentage = (present_days/total_days)*100

    if attendence_percentage <75:
       return f"NOT ELIGIBLE: {attendence_percentage}"
    else :
        return f"ELIGIBLE :  {attendence_percentage}"



attendence = []
for i in range(10):
    print ("day:",i)
    status = input("A/P").upper()
    attendence.append(status)
print(attendence)
result =  employee_attendance_eligibility_checker(attendence)
print(result)
```

# Problem 6: Password Strength Checker


```python
def check_password_strength(password):
    chars = list(password)
    length=len(password)
    print(chars)
    if (length <8):
        return "INVALID PASSWORD: less than 8 chars"
        
    else:
        digit = False
        symbol = False
        for ch in password:
            if ch.isdigit():
                digit = True
            if ch in "@#$":
                symbol = True
        if digit and symbol:
            return "VALID PASSWORD"
        else:
            return "INVALID PASSWORD"

password = input("Enter password")
result = check_password_strength(password)
print(result)
```
