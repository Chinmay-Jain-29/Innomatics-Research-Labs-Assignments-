```python
#Q1

# Stored credentials
username = "admin"
password = "1234"

# Taking user input
input_user = input("Enter Username: ")
input_pass = input("Enter Password: ")

# Checking login
if input_user == username and input_pass == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#Q2

# List of student marks
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
#Q3

# Original list
names = [" Alice ", "bob", " CHARLIE "]

# Cleaned list
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Original List:", names)
print("Cleaned List:", cleaned_names)
```

```python
#Q4

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}'")
    print("Length:", length)
    
    if length > 10:
        print("Status: Longer than 10 characters ⚠")
    else:
        print("Status: Within limit ✅")
    
    print()  # blank line for readability
```

```python
#Q5

# List of system logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count ERROR entries
error_count = logs.count("ERROR")

print("Total ERROR entries:", error_count)
```
