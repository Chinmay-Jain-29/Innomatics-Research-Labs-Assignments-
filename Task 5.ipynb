# Problem 1: Social Media – Post Engagement Analyzer


```python
def post_engagement_analyzer():
    num_users = int(input("Enter number of users who liked the post: "))
    
    total_likes = 0
    
    for i in range(num_users):
        likes = int(input("Enter likes from user: "))
        total_likes += likes
    
    print("Total Likes:", total_likes)
    
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


post_engagement_analyzer()
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_adequacy_checker(record):
    summ = 0
   
    for i in record:
        summ = summ+i
    avg = (summ)/len(record)
    print("AVERAGE RAINFALL= ",avg)
    if avg >=60 and avg<=120:
        status = "NORMAL RAINFALL"
    elif avg>120:
        status = "ADEQUATE RAINFALL"
    else:
        status="INADQUATE RAINFALL"
    return status
    
record= []
num_of_days = int(input("ENTER THE NUMBER OF DAYS"))

for i in range(1,num_of_days+1):
    print("Day:",i)
    rainfall = float(input("ENTER THE RAINFALL IN cm"))
    record.append(rainfall)

result = rainfall_adequacy_checker (record)
print(result)
```

# Problem 4: Social Media – Duplicate Account Detection


```python
def duplicate_account_detection():
    accounts = set()
    i=1
    
    while(i!=0):
        account_name = input("ENTER YOUR ACCPUNT NAME")
        if account_name in accounts:
            status = "YES"
            print("data already found")
        else:
            status = "NO"
            print("VALID NAME: ")
            accounts.add(account_name)

        i=int(input("WANTS TO VERIFY ANOTHER ACCOUNT NAME (1/0)"))
        print("EXITEDD")

duplicate_account_detection()
```

# Problem 6: Agriculture – Premium Crop Price Filter


```python
def premium_crop_filter():
    num_of_crops = int(input("ENTER THE NUMBER OF CROPS"))
    premium_crop = []
    for _ in range(num_of_crops):
        crop_price = int(input("ENTER THE CROP PRICE"))
        if crop_price>=2000:
            premium_crop.append(crop_price)

    return f"premium crops : {premium_crop}"


result=premium_crop_filter()
print(result)
```

# Problem 8: Banking – Daily Transaction Limit Checker


```python
def daily_transaction_limit_checker():
    transactions = []
    summ=0
    i=1
    while(i!=0):
        transaction_amount = int(input("ENTER THE TRANSACTION AMOUNT"))
        if (sum(transactions)+transaction_amount > 50000):
            print("TRANSACTION FAILES:  LIMIT EXCEEDS")
        else:
            transactions.append(transaction_amount)
            print("TRANSACTION SUCCESSFUL")
        print("TODAYS TRANSACTIONS: " ,transactions ,"TOTAL TRANSACTION AMOUNT", sum(transactions))
        i = int(input("wants to perform another transaction?? (1/0)"))
        
daily_transaction_limit_checker()
```

# Problem 10: Smart Electricity Bill Analyzer


```python
def smart_electric_bill_analyzer():
    days = int(input("ENTER THE NUMBER OF DAYS: "))
    
    total_units = 0
    
    for i in range(days):
        units = int(input("Enter the used electric units: "))
        total_units += units
    

    if total_units <= 100:
        bill = total_units * 3
    elif total_units <= 200:
        bill = (100 * 3) + ((total_units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((total_units - 200) * 7)

    print("Total Units:", total_units)
    print("Total Bill: ₹", bill)

    if bill < 500:
        print("Low Usage")
    elif bill <= 1500:
        print("Moderate Usage")
    else:
        print("High Usage")

smart_electric_bill_analyzer()
```
